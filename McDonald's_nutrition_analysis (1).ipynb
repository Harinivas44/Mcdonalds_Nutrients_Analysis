```python
correlation=df[['Calories','Calories from Fat','Cholesterol','Sodium','Carbohydrates','Dietary Fiber','Sugars','Protein']].corr()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(correlation,annot=True)
plt.show()
```

Positive Correlation:

Calories and Calories from Fat: These two variables have a strong positive correlation of approximately 0.905. It means that as the total calories in a food item increase, the calories coming from fat also tend to increase proportionally. This indicates that foods with higher total calorie content are likely to have a higher proportion of calories coming from fat.
Calories and Protein: There is a strong positive correlation of about 0.788 between calories and protein. It suggests that foods with higher calorie content are likely to be richer in protein as well.Where the Calories and Cholesterol also have positive correlation of 0.6 .

Negative Correlation:

Calories from Fat and Dietary Fiber: These two variables have a negative correlation of approximately -0.295. This implies that foods with a higher proportion of calories coming from fat are likely to have lower dietary fiber content, and vice versa.
Sugars and Calories from Fat: There is a negative correlation of about -0.115 between sugars and calories from fat. It suggests that foods with higher sugar content are less likely to have a high proportion of calories coming from fat.
Sugars and Dietary Fiber: These two variables are negatively correlated with a value of approximately -0.295. It means that foods with higher sugar content tend to have lower dietary fiber content, and vice versa.

Zero Correlation:

Calories and Sugars: These two variables have a correlation coefficient of approximately 0.260, which is close to zero. It indicates a weak correlation between total calories and sugar content. In other words, the total calorie content of a food item is not strongly influenced by its sugar content.
Calories from Fat and Sugars: The correlation coefficient between calories from fat and sugars is -0.115, which is also close to zero. It suggests a weak relationship between the proportion of calories coming from fat and sugar content in a food item.

```python
nutrients = ['Calories', 'Calories from Fat', 'Cholesterol', 'Sodium', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein']
data_list = []
for i in nutrients:
    max_value = max(df[i])
    max_item = df[df[i] == max_value]["Item"].iloc[0]
    data_type = df[i].dtype
    data_list.append({'Nutrient': i, 'Food Item': max_item, 'Value': max_value})
result_df = pd.DataFrame(data_list)
result_df
```

### The Food Items which have less nutrients !!!

```python
nutrients = ['Calories', 'Calories from Fat', 'Cholesterol', 'Sodium', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein']
data_list = []
for i in nutrients:
    min_value = min(df[i])
    min_item = df[df[i] == min_value]["Item"].iloc[0]
    data_type = df[i].dtype
    data_list.append({'Nutrient': i, 'Food Item': min_item, 'Value': min_value})
result_df = pd.DataFrame(data_list)
result_df
```

```python
fig, axes = plt.subplots(2, 2, figsize=(25, 15))

sns.regplot(x='Calories', y='Cholesterol', data=df, ax=axes[0, 0])
axes[0, 0].set_xlabel('Calories', fontsize=15)
axes[0, 0].set_ylabel('Cholesterol', fontsize=15)
axes[0, 0].tick_params(axis='both', which='major', labelsize=15)
axes[0, 0].set_title('Calories vs. Cholesterol', fontsize=18)

sns.regplot(x='Calories', y='Protein', data=df, ax=axes[0, 1])
axes[0, 1].set_xlabel('Calories', fontsize=15)
axes[0, 1].set_ylabel('Protein', fontsize=15)
axes[0, 1].tick_params(axis='both', which='major', labelsize=1)
axes[0, 1].set_title('Calories vs. Protein', fontsize=18)

sns.regplot(x='Calories', y='Calories from Fat', data=df, ax=axes[1, 0])
axes[1, 0].set_xlabel('Calories', fontsize=15)
axes[1, 0].set_ylabel('Calories from Fat', fontsize=15)
axes[1, 0].tick_params(axis='both', which='major', labelsize=15)
axes[1, 0].set_title('Calories vs. Calories From Fat', fontsize=18)

sns.regplot(x='Calories', y='Carbohydrates', data=df, ax=axes[1, 1])
axes[1, 1].set_xlabel('Calories', fontsize=15)
axes[1, 1].set_ylabel('Carbohydrates', fontsize=15)
axes[1, 1].tick_params(axis='both', which='major', labelsize=15)
axes[1, 1].set_title('Calories vs. Carbohydrates', fontsize=18)

plt.show()
```

```python
nutrients = ['Calories', 'Calories from Fat', 'Cholesterol', 'Sodium', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein']
colors = ['brown', 'orange', 'red', 'violet', 'yellow', 'pink', 'gray', 'green']


plt.figure(figsize=(25,45))
for i in range(len(nutrients)):
    plt.subplot(4, 2, i + 1)
    cat = df.sort_values(nutrients[i], ascending=False).head()
    cat['Item']=cat['Item'].apply(lambda x:'\n'.join(x.split()))
    sns.barplot(x=nutrients[i], y='Item', data=cat, color=colors[i],errorbar=('ci',300))
    plt.title(f'TOP 5 Items based on {nutrients[i]}(Nutrient-rich)',fontsize=20)
    plt.xlabel(nutrients[i],fontsize=20)
    plt.ylabel('Items',fontsize=20)
    plt.xticks(fontsize=16)  
    plt.yticks(fontsize=16)
plt.tight_layout()
plt.show()
```
